```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get("https://www.opinet.co.kr/searRgSelect.do")
time.sleep(2)

for i in range(1, 8):
    xpath = f'//*[@id="body1"]/tr[{i}]/td[1]/a'

    try:
        # 버튼 클릭 전 재탐색
        button = driver.find_element(By.XPATH, xpath)
        button.click()

        time.sleep(1)  # os_nm 로딩 기다림

        # 클릭 후 요소 새로 가져오기
        elements = driver.find_elements(By.XPATH, '//*[@id="os_nm"]')

        if not elements:
            print(f"[{i}] os_nm 없음")
            continue

        for el in elements:
            try:
                print(f"[{i}] 주유소 이름:", el.text.strip())
            except Exception as e:
                print(f"[{i}] 요소 접근 실패 (stale):", e)

    except Exception as e:
        print(f"[{i}] 오류 발생:", e)
        time.sleep(1)

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
import csv

# 크롬 드라이버 실행
driver = webdriver.Chrome()
driver.get("https://www.opinet.co.kr/searRgSelect.do")
time.sleep(2)

# 결과 저장용 리스트
data_list = []

for i in range(1, 8):
    xpath = f'//*[@id="body1"]/tr[{i}]/td[1]/a'
    try:
        # 링크 클릭
        button = driver.find_element(By.XPATH, xpath)
        button.click()
        time.sleep(1.5)  # 상세 정보 로딩 대기

        # 정보 수집
        # name = driver.find_element(By.XPATH, '//*[@id="os_nm"]').text.strip()
        # address = driver.find_element(By.XPATH, '//*[@id="rd_addr"]').text.strip()
        # brand = driver.find_element(By.XPATH, '//*[@id="poll_div_nm"]').text.strip()
        name = driver.find_element(By.ID, "os_nm").get_attribute("innerText").strip()
        address = driver.find_element(By.ID, "rd_addr").get_attribute("innerText").strip()
        brand = driver.find_element(By.ID, "poll_div_nm").get_attribute("innerText").strip()
        gasolin_price = driver.find_element(By.ID, "b027_p").get_attribute("innerText").strip()
        diesel_price = driver.find_element(By.ID, "d047_p").get_attribute("innerText").strip()


        print(f"[{i}] 이름: {name}, 주소: {address}, 브랜드: {brand}, 가솔린: {gasolin_price}, 디젤: {diesel_price}")

        data_list.append({
            "이름": name,
            "주소": address,
            "브랜드": brand,
            "가솔린": gasolin_price,
            "디젤": diesel_price
        })

    except Exception as e:
        print(f"[{i}] 오류 발생:", e)
        time.sleep(1)

# 브라우저 종료
driver.quit()

# ✅ CSV 저장
filename = "주유소정보.csv"
with open(filename, "w", encoding="utf-8-sig", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["이름", "주소", "브랜드", "가솔린", "디젤"])
    writer.writeheader()
    writer.writerows(data_list)

print(f"\n✅ 수집 완료! CSV 파일로 저장되었습니다 → {filename}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select

import time
import csv

gu_list = ["강남구", "강동구", "강북구", "강서구", "관악구", "광진구", "구로구", "금천구", "노원구", "도봉구", "동대문구", "동작구", "마포구", "서대문구", "서초구", "성동구", "성북구", "송파구", "양천구", "영등포구", "용산구", "은평구", "종로구", "중구", "중랑구"]

driver = webdriver.Chrome()
driver.get("https://www.opinet.co.kr/searRgSelect.do")
time.sleep(2)

data_list = []

for gu in gu_list:
    try:    
        dropdown = Select(driver.find_element(By.ID, "SIGUNGU_NM0"))
        dropdown.select_by_visible_text(gu)
        print(f"[{gu}] 선택됨")

        time.sleep(2)

        rows = driver.find_elements(By.XPATH, '//*[@id="body1"]/tr')
        print(f"[{gu}] 총 {len(rows)}개의 주유소가 있습니다.")

        for i in range(1, len(rows) + 1):
            xpath = f'//*[@id="body1"]/tr[{i}]/td[1]/a'
            try:
                button = driver.find_element(By.XPATH, xpath)
                button.click()
                time.sleep(1.5)

                name = driver.find_element(By.ID, "os_nm").get_attribute("innerText").strip()
                address = driver.find_element(By.ID, "rd_addr").get_attribute("innerText").strip()
                brand = driver.find_element(By.ID, "poll_div_nm").get_attribute("innerText").strip()
                gasolin_price = driver.find_element(By.ID, "b027_p").get_attribute("innerText").strip()
                diesel_price = driver.find_element(By.ID, "d047_p").get_attribute("innerText").strip()

                # print(f"  - 이름: {name}, 주소: {address}, 브랜드: {brand}, 휘발유: {gasolin_price}, 경유: {diesel_price}")
                data_list.append({
                        "지역": gu,
                        "이름": name,
                        "주소": address,
                        "브랜드": brand,
                        "휘발유": gasolin_price,
                        "경유": diesel_price
                    })

            except Exception as e:
                print(f"[{i}] 오류 발생:", e)
                continue
    except Exception as e:
        print(f"[{gu}] 오류 발생:", e)
        continue

driver.quit()

filename = "서울시주유소정보.csv"
with open(filename, "w", encoding="utf-8-sig", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["지역", "이름", "주소", "브랜드", "가솔린", "디젤"])
    writer.writeheader()
    writer.writerows(data_list)

print(f"\n 데이터 수집 완료! CSV 파일로 저장되었습니다 → {filename}")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time

driver = webdriver.Chrome()
driver.get("https://www.opinet.co.kr/searRgSelect.do")
time.sleep(2)

# tr 개수 확인
rows = driver.find_elements(By.XPATH, '//*[@id="body1"]/tr')
print(f"총 {len(rows)}개의 행이 있습니다.")

for i in range(2, len(rows) + 1):
    xpath = f'//*[@id="body1"]/tr[{i}]/td[1]/a'
    try:
        # a 태그가 없는 행은 패스
        link = driver.find_element(By.XPATH, xpath)

        if not link.text.strip():
            print(f"[{i}] a 태그는 있지만 텍스트 없음. 건너뜀.")
            continue

        link.click()
        time.sleep(1.5)

        name = driver.find_element(By.XPATH, '//*[@id="os_nm"]')
        print(f"[{i}] 주유소 이름:", name.text)

        driver.back()
        time.sleep(1.5)

    except Exception as e:
        print(f"[{i}] 오류 발생 (a 태그 없음 등):", e)
        driver.get("https://www.opinet.co.kr/searRgSelect.do")
        time.sleep(2)

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.get("https://www.opinet.co.kr/searRgSelect.do")
time.sleep(3)

# 1. 전체 행 수 파악
rows = driver.find_elements(By.CSS_SELECTOR, "#body1 > tr")
print(f"총 {len(rows)}개의 행이 있습니다.")

# 2. 행 수만큼 반복
for i in range(1, len(rows) + 1):  # nth-child는 1부터 시작
    selector = f'#body1 > tr:nth-child({i}) > td.rlist > a'

    try:
        link = driver.find_element(By.CSS_SELECTOR, selector)

        # 링크 클릭
        link.click()
        time.sleep(1)

        # 상세 정보 추출
        os_nm = driver.find_element(By.ID, "os_nm")
        print(f"[{i}] 주유소 이름:", os_nm.text)

        # 목록 페이지로 되돌아가기
        driver.back()
        time.sleep(1.5)

    except Exception as e:
        print(f"[{i}] 오류 발생 또는 a 태그 없음:", e)
        driver.get("https://www.opinet.co.kr/searRgSelect.do")
        time.sleep(2)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Chrome()
driver.get("https://www.opinet.co.kr/searRgSelect.do")

wait = WebDriverWait(driver, 10)

# 목록 페이지 로딩 기다리기
wait.until(EC.presence_of_element_located((By.XPATH, '//*[@id="body1"]/tr')))

# 행 개수 세기
rows = driver.find_elements(By.XPATH, '//*[@id="body1"]/tr')
row_count = len(rows)
print(f"총 {row_count}개의 행이 있습니다.")

for i in range(1, row_count + 1):
    xpath = f'//*[@id="body1"]/tr[{i}]/td[1]/a'

    try:
        link = driver.find_element(By.XPATH, xpath)
        if not link:
            continue

        # 클릭
        link.click()

        os_nm = driver.find_element(By.ID, "os_nm").text.strip()
        print(f"[{i}] 주유소 이름:", os_nm)

        time.sleep(1)

    except Exception as e:
        print(f"[{i}] 에러:", e)
        continue
```
